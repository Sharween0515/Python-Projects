```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
theme.head()
```

```python
df.head()
```

```python
merged = df.merge(theme, left_on="parent_theme",right_on="name")
```

```python
merged = merged.dropna(subset=["set_num"])
```

## Percentage of lego sets that are Star Wars and licensed 

```python
star_wars = merged[merged["is_licensed"] & merged["theme_name"].str.contains("Star Wars")]
```

```python
star_wars.head()
```

```python
swc = star_wars.count()["set_num"].astype("int32")
```

```python
dfc = df.count()["set_num"].astype("int32")
```

```python
percentage = (swc / dfc * 100).astype("float").round(2)
```

```python
percentage
```

## Year Star Wars was not the most popular theme among licensed themes

```python
licensed = merged[merged["is_licensed"]]
```

```python
licensed = licensed.sort_values("year")
```

```python
sorted = licensed.groupby(["year","parent_theme"])[["is_licensed"]].sum().reset_index()
```

```python
df_sorted = sorted.sort_values("is_licensed",ascending=False).drop_duplicates(["year"])
```

```python
df_sorted = df_sorted.sort_values(["year"])
```
